```python
import csv
f = open('범죄율_2016.csv')

name = input('알고 싶은 지역구를 입력하세요:')
data = csv.reader(f)
next(data)
data = list(data)
crime = []
find = []

for row in data :
    if row[0] == name and row[2] == '발생' :
        crime.append(int(row[3]))
    elif row[0] == name and row[2] == '검거' :
        find.append(int(row[3]))

import matplotlib.pyplot as plt

plt.figure(figsize=(6,4), dpi=500)
plt.rc('font',family='Malgun Gothic')

crimegraph = ['살인','강도','강간','절도','폭력']
occurence = [crime[0],crime[1],crime[2],crime[3],crime[4]]
catch = [find[0],find[1],find[2],find[3],find[4]]

plt.bar(range(0,9,2),occurence, width = 0.4, label = '발생횟수')
plt.bar(range(1,10,2),catch, width = 0.4, label = '검거횟수')
labels = '살인','강도','강간','절도','폭력'

plt.title('입력하신 '+name+'구의 범죄 현황입니다.')

plt.legend()
plt.xticks([0.5,2.5,4.5,6.5,8.5],['살인','강도','강간','절도','폭력'])
```

```python
import csv
f = open('범죄율_2016.csv')

data = csv.reader(f)
next(data)
data = list(data)
high = 0
local = ''
add_crime = 0
add_find = 0
loc = '중부'
for row in data :
    if row[0] == loc :
        if row[2] == '발생' :
            add_crime = add_crime + int(row[3])
        elif row[2] == '검거' :
            add_find = add_find + int(row[3])
    
    else :
        rate = float(add_find/add_crime)
        if rate>high :
            high = rate
            local = loc
        loc = row[0]
        add_crime = 0
        add_find = 0
            
print(local)
print(high)
```
